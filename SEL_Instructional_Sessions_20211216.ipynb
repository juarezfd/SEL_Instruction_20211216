```python
import pandas as pd
import altair as alt
import numpy as np
from datetime import datetime as dt
```

```python
SEL_Instruction = pd.read_csv("SEL_Instruction.csv") 
#load csv of Vanderbilt course code, subject area, department, and school for the
# classes Stevenson Engineering Library (SEL) has thought since Honora joined in Fiscal Year 2018
```

```python
SEL_Instruction
```

```python
SEL_Instruction['Code'].sort_values().unique(), SEL_Instruction['Code'].sort_values().nunique()
```

```python
SEL_Instruction['Code'] = SEL_Instruction['Code'].apply(lambda x: x.strip()) 
# cleaning the Code column since we have things like 'CHEM' and 'CHEM ' that are considered different
```

```python
SEL_Instruction['Code'].sort_values().unique(), SEL_Instruction['Code'].sort_values().nunique()
```

```python
Code_Data_Dictionary = pd.read_csv("SEL_Instruction_Code_Data_Dictionary.csv")
```

```python
JoinDf = pd.merge(SEL_Instruction, Code_Data_Dictionary, on='Code', how='inner')
# Combines two dataframes on a key/columns in this case the Code 
# We are doing this so we can learn which classes SEL thought were engineering versus others
```

```python
FY18 = JoinDf[JoinDf['Fiscal_Year'] == 18] # fall 2018 dataframe info only
FY22 = JoinDf[JoinDf['Fiscal_Year'] == 22] # fall 2022 dataframe info only
```

```python
JoinDf['Engineering_Class'] = np.where(JoinDf['School'] == 'School of Engineering', True, False)
# adding a column to let us know which are Instruction sessions we did for engineering classes 
```

```python
df = pd.read_csv("Honoras_Engineering_Meeting_Tidy.csv") 
```

```python
bars = alt.Chart(df, title='SEL Instructional Session Summary').mark_bar(opacity=0.8).encode(
    x=alt.X('Academic Year:O'),
    y=alt.Y('count(Academic Year):Q',title="Count of Instructional Sessions"),
    color='Engineering_Other:N',
    tooltip=['count(Academic Year):Q']
).configure_axis(
    labelFontSize=20,
    titleFontSize=20
)



(bars).properties(height=500, width=600).configure_title(fontSize=20)

```
